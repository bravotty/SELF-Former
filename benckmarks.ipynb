```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['font.sans-serif'] = ['Times New Roman']
del matplotlib.font_manager.weight_dict['roman']

df0 = pd.read_csv(r'data_vis2/metrics_ours.csv')
data0 = df0.iloc[:, 1]
df1 = pd.read_csv(r'data_vis2/metrics_Tangram.csv')
data1 = df1.iloc[:, 1]
df2 = pd.read_csv(r'data_vis2/metrics_novoSpaRc.csv')
data2 = df2.iloc[:, 1]
df3 = pd.read_csv(r'data_vis2/metrics_GraphST.csv')
data3 = df3.iloc[:, 1]
df4 = pd.read_csv(r'data_vis2/metrics_SpaOTsc.csv')
data4 = df4.iloc[:, 1]
df5 = pd.read_csv(r'data_vis2/metrics_STEM.csv')
data5 = df5.iloc[:, 1]
df6 = pd.read_csv(r'data_vis2/metrics_gimVI.csv')
data6 = df6.iloc[:, 1]
df7 = pd.read_csv(r'data_vis2/metrics_stDiff.csv')
data7 = df7.iloc[:, 1]
final = pd.DataFrame({"Ours": data0, "Tangram": data1, "novoSpaRc": data2, "GraphST": data3, "SpaOTsc": data4, "STEM": data5, "gimVI": data6, "stDiff": data7})
df_melted = final.melt(var_name='Method', value_name='Value')
plt.figure(figsize=(10, 6))
palette = sns.color_palette("pastel")
boxplot = sns.boxplot(
    x='Value', 
    y='Method', 
    data=df_melted, 
    palette=palette, 
    linewidth=2.5,
    boxprops=dict(alpha=0.85),
    whiskerprops=dict(alpha=0.85),
    capprops=dict(alpha=0.85),
    medianprops=dict(alpha=0.85),
    flierprops=dict(markerfacecolor='r', marker='o', markersize=3, alpha=0.5)
)
plt.yticks(rotation=45, fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

```python
# RMSE
df0 = pd.read_csv(r'data_vis2/metrics_ours.csv')
data0 = df0.iloc[:, 4]
df1 = pd.read_csv(r'data_vis2/metrics_Tangram.csv')
data1 = df1.iloc[:, 4]
df2 = pd.read_csv(r'data_vis2/metrics_novoSpaRc.csv')
data2 = df2.iloc[:, 4]
df3 = pd.read_csv(r'data_vis2/metrics_GraphST.csv')
data3 = df3.iloc[:, 4]
df4 = pd.read_csv(r'data_vis2/metrics_SpaOTsc.csv')
data4 = df4.iloc[:, 4]
df5 = pd.read_csv(r'data_vis2/metrics_STEM.csv')
data5 = df5.iloc[:, 4]
df6 = pd.read_csv(r'data_vis2/metrics_gimVI.csv')
data6 = df6.iloc[:, 4]
df7 = pd.read_csv(r'data_vis2/metrics_stDiff.csv')
data7 = df7.iloc[:, 4]
final = pd.DataFrame({"Ours": data0, "Tangram": data1, "novoSpaRc": data2, "GraphST": data3, "SpaOTsc": data4, "STEM": data5, "gimVI": data6, "stDiff": data7})
df_melted = final.melt(var_name='Method', value_name='Value')
df_melted = df_melted.dropna(subset=['Method', 'Value'])
df_summary = df_melted.groupby('Method').agg(['mean', 'sem']).reset_index()
df_summary.columns = ['Method', 'Mean', 'SEM']
method_order = ['Ours', 'Tangram', 'novoSpaRc', 'GraphST', 'SpaOTsc', 'STEM', 'gimVI', 'stDiff']
sns.set_theme(style="whitegrid")
palette = sns.color_palette("pastel")
min_method = df_summary.loc[df_summary['Mean'].idxmin()]
plt.figure(figsize=(12, 8))
plt.gca().patch.set_edgecolor('black')
plt.gca().patch.set_linewidth(1)
barplot = sns.barplot(
    x='Mean', 
    y='Method', 
    data=df_summary, 
    palette=palette, 
    ci=None,
    xerr=df_summary['SEM'],
    order=method_order
)

plt.yticks(rotation=45, fontsize=15)
plt.xticks(fontsize=15)
sns.despine(left=True, bottom=True)
plt.grid(True, alpha=0.5)
barplot.plot(1.17, 0, "*", markersize=19, color="r")
plt.show()



```
